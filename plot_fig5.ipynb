```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
disruption_f = np.concatenate((np.ones(30) * 100, np.ones(25) * 0.4 * 100, np.ones(55) * 100))

fig, axs = plt.subplots(1, 1, figsize=(6, 6))

#axs.plot(topbest_prod[:, i], label='Highest 1â€°')
axs.plot(disruption_f, label='Real Production Volume')

#axs.plot(disaster_prod[:, i], label='Lowest sample')
axs.tick_params(axis='y', labelsize=14)
axs.tick_params(axis='x', labelsize=14)

#axs.arrow(0, 0, 20, 1, head_width=0.001)
axs.annotate(text='$\\tau_{r}$', xy=(30,60), xytext=(55,58.5), arrowprops=dict(arrowstyle='<->'), fontsize=16)
axs.annotate(text='$c_{r}$', xy=(38,100), xytext=(35,35), arrowprops=dict(arrowstyle='<->'), fontsize=16)
axs.annotate(text='', xy=(31,100), xytext=(41,100), arrowprops=dict(arrowstyle='-'), fontsize=16)
axs.annotate(text='risk event $p_r$ ', xy=(29,100), xytext=(11,120), arrowprops=dict(arrowstyle='->'), fontsize=16)

axs.set_ylim(0.0, 130)
axs.legend(loc='lower right', fontsize=14)
axs.set_ylabel('Production % of Max', fontsize=14)
axs.set_xlabel('Days', fontsize=14)



plt.show()
```
